```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
output = pd.read_csv("exp6_T95_minus_UT95.csv")
```

```python
output.set_index(["tau", "crossImmu", "emergence", "location", "treatment", "Bp", "Bg"]).sort_index().name.tolist()
```

```python
# random = output[(output.treatment == "random")]
reshaped = output[["treatment", "Bp", "Bg", "crossImmu", "tau", "emergence", "location", "med"]].pivot_table(index = ["tau", "crossImmu", "emergence", "location"], columns = ["treatment", "Bp", "Bg"], values="med", aggfunc=np.median) #.set_index(["treatment", "Bg", "Bp"])
sns.heatmap(data=reshaped)
```
